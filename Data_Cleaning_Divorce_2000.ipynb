```python
mar_div_10_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2000 County-Level-Divorce-Data-Spreadsheet.csv")
mar_div_10_df.fillna(value=0, inplace=True)
mar_div_10_df.head(10)
```

```python
mar_div_10_df['County'] = mar_div_10_df['County'].str.replace(" ", "")
mar_div_10_df['Geographic Area'] = mar_div_10_df['County'] + ", " + mar_div_10_df['State']

#mar_div_10_df.dtypes
mar_div_10_df.head(10)
```

```python
mar_div_10_df.drop(columns=['County', 'State', 'Married_Men', 'Married_Women', 'Crude_Divorce_Rate',	'Adjusted_Divorce_Rate' ], inplace=True)
mar_div_10_df.head(10)
```

```python
mar_div_10_df['FIPS Code'] = mar_div_10_df['FIPS Code'].astype('str').str.zfill(5)
mar_div_10_df.head(10)
```

```python
mar_div_10_df['year'] = 2000
mar_div_10_df.head(10)
```

```python

dict = {'FIPS Code': 'co_fips',
        'Population': 'Census',
        'Number_of_Divorces': 'divorces_2000',
        'Total_Married': 'marriages_2000'
              
        }
mar_div_10_df.rename(columns=dict,
          inplace=True)

mar_div_10_df = mar_div_10_df[['co_fips', 'divorces_2000', 'marriages_2000', 'Geographic Area', 'Census', 'year']]

display(mar_div_10_df)
```
