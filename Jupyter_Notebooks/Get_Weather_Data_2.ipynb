```python
import pandas as pd
from config import NOAA_API_KEY
import requests
import json

print(NOAA_API_KEY[1])
```

```python
#altered url to reduce number of requests
#api_url_base = f'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GSOY&datatypeid=TAVG&datatypeid=DX70&datatypeid=DT32&locationid=FIPS:12017&units=standard&startdate=2001-01-01&enddate=2011-01-01', headers={'token':NOAA_API_KEY})


#Empty list for appending data
county_temp_data = []
county_precip_data = []
#Lists to loop through
from county_identifiers import FIPS_List
temp_datatypes = ['TAVG','DX90','DT32']
precip_datatypes = []
years = [2001,2011]

#Counter for number of api requests
i = 0

#Index for api key list
k=0
```

```python
#API Requests to get temperature data per county

for fips in FIPS_List:
    for year in years:
        for datatype in temp_datatypes:
            i = i + 1
            if i > 9850:
               k = k + 1
               i = 0 
            try:
                county_weather = requests.get(f'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GSOY&datatypeid={datatype}&locationid=FIPS:{fips}&units=standard&startdate={year}-01-01&enddate={year}-01-01', headers={'token':NOAA_API_KEY[k]}).json()
                        
                if datatype == 'TAVG':
                    temp_avg = county_weather['results'][0]['value']
                elif datatype == 'DX90':
                    days_over_70 = county_weather['results'][0]['value']
                else:
                    days_under_32 = county_weather['results'][0]['value']
                
                county_temp_data.append({"Fips Code": fips,
                                            "Year": year,
                                            "Average Temp": temp_avg,
                                            "Days Over 70": days_over_70,
                                            "Days Under 32": days_under_32})
                                            
                print(f"Getting Data for FIPS:{fips},{datatype}, and {year}")
            except:
                print(f"{fips} data {datatype} not found for {year}. Skipping...")
                pass

    
            
```

```python
county_temp_data_df = pd.DataFrame(county_temp_data)
county_temp_data_df.head()
```

```python
# Create the output file (CSV).
output_data_file = "../Data/county_temps.csv"
# Export the City_Data into a CSV.
county_temp_data_df.to_csv(output_data_file, index=False)
```
