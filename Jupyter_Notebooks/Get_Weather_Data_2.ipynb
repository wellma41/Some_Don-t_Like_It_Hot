```python
from county_identifiers import FIPS_List
county_data_weather = []
datatypes = ['TAVG','DX70','DT32']
years = [2001,2011]
for fips in FIPS_List:
    for year in years:
        for datatype in datatypes:

            try:
                county_weather = requests.get(f'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GSOY&datatypeid={datatype}&locationid=FIPS:{fips}&units=standard&startdate={year}-01-01&enddate={year}-01-01', headers={'token':NOAA_API_KEY}).json()
                        
                if datatype == 'TAVG':
                    temp_avg = county_weather['results'][0]['value']
                elif datatype == 'DX70':
                    days_over_70 = county_weather['results'][0]['value']
                else:
                    days_under_32 = county_weather['results'][0]['value']
                
                county_data_weather.append({"Fips Code": fips,
                                            "Year": year,
                                            "Average Temp": temp_avg,
                                            "Days Over 70": days_over_70,
                                            "Days Under 32": days_under_32})
                print(f"Getting Data for FIPS:{fips},{datatype}, and {year}")
            except:
                print(f"{fips} data {datatype} not found for {year}. Skipping...")
                pass

    
            
```
