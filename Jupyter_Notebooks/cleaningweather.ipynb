```python
import pandas as pd
import os
```

```python
precip_df = pd.read_csv("../Data/county_precip.csv")
temp_df = pd.read_csv("../Data/county_temps.csv")
precip_df
```

```python
# precip_df.fillna(value=0, inplace=True)
temp_df = temp_df.fillna(value=0, inplace=True)
temp_df.head(10)
```

```python
dfp_2001 = precip_df.loc[precip_df['Year'] == 2001]
dfp_2011 = precip_df.loc[precip_df['Year'] == 2011]
# dft_2001 = temp_df.loc[temp_df['Year'] == 2001]
# dft_2011 = temp_df.loc[temp_df['Year'] == 2011]
dfp_2001.head()
```

```python
did_it_rain_avg = dfp_2001.groupby(["Fips Code"]).mean()["Days with > 0.1 inch of Precipitation"]

extreme_max_avg = dfp_2001.groupby(["Fips Code"]).mean()["Extreme maximum precipitation"]

oneinchofsnow_avg = dfp_2001.groupby(["Fips Code"]).mean()["Days with snow depth > 1 inch"]
```

```python
county_precip_data_clean_01=pd.DataFrame({
    "Year": 2001,
    "Days with > 0.1 inch of Precipitation": did_it_rain_avg,
    "Extreme maximum precipitation": extreme_max_avg,
    "Days with snow depth > 1 inch": oneinchofsnow_avg})
county_precip_data_clean_01.head()        
```

```python
county_precip_data_clean_01["Fips Code"]= county_precip_data_clean_01.index
county_precip_data_clean_01.head()  
```

```python
did_it_rain_avg = dfp_2011.groupby(["Fips Code"]).mean()["Days with > 0.1 inch of Precipitation"]

extreme_max_avg = dfp_2011.groupby(["Fips Code"]).mean()["Extreme maximum precipitation"]

oneinchofsnow_avg = dfp_2011.groupby(["Fips Code"]).mean()["Days with snow depth > 1 inch"]
```

```python
county_precip_data_clean_11=pd.DataFrame({ 
    "Year":2011,                            
    "Days with > 0.1 inch of Precipitation": did_it_rain_avg,
    "Extreme maximum precipitation": extreme_max_avg,
    "Days with snow depth > 1 inch": oneinchofsnow_avg})
county_precip_data_clean_11.head()   
```

```python
county_precip_data_clean_11["Fips Code"]= county_precip_data_clean_11.index
county_precip_data_clean_11.head()
```

```python
df_precip_avg = county_precip_data_clean_01.append(county_precip_data_clean_11, ignore_index=True)

df_precip_avg.head()
```

```python
df_precip_avg['Fips Code'] = df_precip_avg['Fips Code'].astype(str).str.zfill(5)
df_precip_avg.head()
```

```python
# Create the output file (CSV).
output_data_file_two = "../Data/avg_county_precip.csv"
# Export the City_Data into a CSV.
df_precip_avg.to_csv(output_data_file_two, index=False)
    
```
