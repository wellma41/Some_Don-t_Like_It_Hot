```python
import pandas as pd
import os
```

```python
precip_df = pd.read_csv("../Data/county_precip.csv")

precip_df
```

```python
dfp_2001 = precip_df.loc[precip_df['Year'] == 2001]
dfp_2011 = precip_df.loc[precip_df['Year'] == 2011]

dfp_2001.head()
```

```python
did_it_rain_avg = dfp_2001.groupby(["Fips Code"]).mean()["Days with > 0.1 inch of Precipitation"]

extreme_max_avg = dfp_2001.groupby(["Fips Code"]).mean()["Extreme maximum precipitation"]

oneinchofsnow_avg = dfp_2001.groupby(["Fips Code"]).mean()["Days with snow depth > 1 inch"]
```

```python
county_precip_data_clean_01=pd.DataFrame({
    "Year": 2000,
    "Days with > 0.1 inch of Precipitation": did_it_rain_avg,
    "Extreme maximum precipitation": extreme_max_avg,
    "Days with snow depth > 1 inch": oneinchofsnow_avg})
county_precip_data_clean_01.head()        
```

```python
county_precip_data_clean_01["Fips Code"]= county_precip_data_clean_01.index
county_precip_data_clean_01.head()  
```

```python
did_it_rain_avg = dfp_2011.groupby(["Fips Code"]).mean()["Days with > 0.1 inch of Precipitation"]

extreme_max_avg = dfp_2011.groupby(["Fips Code"]).mean()["Extreme maximum precipitation"]

oneinchofsnow_avg = dfp_2011.groupby(["Fips Code"]).mean()["Days with snow depth > 1 inch"]
```

```python
county_precip_data_clean_11=pd.DataFrame({ 
    "Year":2010,                            
    "Days with > 0.1 inch of Precipitation": did_it_rain_avg,
    "Extreme maximum precipitation": extreme_max_avg,
    "Days with snow depth > 1 inch": oneinchofsnow_avg})
county_precip_data_clean_11.head()   
```

```python
county_precip_data_clean_11["Fips Code"]= county_precip_data_clean_11.index
county_precip_data_clean_11.head()
```

```python
df_precip_avg = county_precip_data_clean_01.append(county_precip_data_clean_11, ignore_index=True)

df_precip_avg.head()
```

```python
df_precip_avg.tail()
```

```python
df_precip_avg['Fips Code'] = df_precip_avg['Fips Code'].astype(str).str.zfill(5)
df_precip_avg.head()
```

```python
# Create the output file (CSV).
output_data_file_two = "../Data/avg_county_precip.csv"
# Export the City_Data into a CSV.
df_precip_avg.to_csv(output_data_file_two, index=False)
    
```

```python
temp_df = pd.read_csv("../Data/county_temps.csv")
temp_df
```

```python
dft_2001 = temp_df.loc[temp_df['Year'] == 2001]
dft_2011 = temp_df.loc[temp_df['Year'] == 2011]
```

```python
temp_avg_01 = dft_2001.groupby(["Fips Code"]).mean()["Average Temp"]

over_90thistime_01 = dft_2001.groupby(["Fips Code"]).mean()["Days Over 70"]

daysunder32_01 = dft_2001.groupby(["Fips Code"]).mean()["Days Under 32"]
```

```python
county_temp_data_clean_01=pd.DataFrame({
    "Year": 2000,
    "Average Temp": temp_avg_01,
    "Days Over 90": over_90thistime_01,
    "Days Under 32": daysunder32_01})

county_temp_data_clean_01["Fips Code"]= county_temp_data_clean_01.index
county_temp_data_clean_01.head()   
```

```python
temp_avg_11 = dft_2011.groupby(["Fips Code"]).mean()["Average Temp"]

over_90thistime_11 = dft_2011.groupby(["Fips Code"]).mean()["Days Over 70"]

daysunder32_11 = dft_2011.groupby(["Fips Code"]).mean()["Days Under 32"]
```

```python
county_temp_data_clean_11=pd.DataFrame({
    "Year": 2010,
    "Average Temp": temp_avg_11,
    "Days Over 90": over_90thistime_11,
    "Days Under 32": daysunder32_11})

county_temp_data_clean_11["Fips Code"]= county_temp_data_clean_11.index
county_temp_data_clean_11.head()    
```

```python
df_temp_avg = county_temp_data_clean_01.append(county_temp_data_clean_11, ignore_index=True)
df_temp_avg['Fips Code'] = df_temp_avg['Fips Code'].astype(str).str.zfill(5)
df_temp_avg.head()
```

```python
df_temp_avg.tail()
```

```python
# Create the output file (CSV).
output_data_file_two = "../Data/avg_county_temp.csv"
# Export the City_Data into a CSV.
df_temp_avg.to_csv(output_data_file_two, index=False)
```

```python
df_precip_avg = df_precip_avg[['Fips Code','Year','Days with > 0.1 inch of Precipitation','Extreme maximum precipitation','Days with snow depth > 1 inch']]
df_temp_avg = df_temp_avg[['Fips Code','Year','Average Temp','Days Over 90','Days Under 32']]
df_temp_avg_sql = df_temp_avg.rename(columns = {'Fips Code' : 'co_fips'})
df_precip_avg_sql= df_precip_avg.rename(columns = {'Fips Code' : 'co_fips'})
```

```python
from sqlalchemy import create_engine
import psycopg2 as psql  
import config as cfg
 
conn_string = f'postgresql://{cfg.PG4_USER}:{cfg.PG4_PWD}@{cfg.PG4_HOST}/{cfg.PG4_DBNAME}'
  
db = create_engine(conn_string)
conn = db.connect()

df_temp_avg_sql.to_sql('County_Temp', con = conn, if_exists = 'replace', index = True)

df_precip_avg_sql.to_sql('County_Precip', con = conn, if_exists = 'replace', index = True)

conn.close()
```
